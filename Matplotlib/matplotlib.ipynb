```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = np.random.random(50)*100
y_data = np.random.random(50)*100
```

Scatter plot

```python
plt.scatter(x_data, y_data)
```

```python
print(x_data)
```

```python
print(y_data)
```

```python
plt.scatter(x_data, y_data, c="red", s=150, marker="o", alpha=0.3)
```

Line plot

```python
years = [2006 + x for x in range(16)]
weights = [20, 30, 68, 79, 98, 19, 49, 60, 67, 70, 79, 100, 45, 55, 78, 80]
```

```python
print(len(years))
print(len(weights))
```

```python
plt.plot(years, weights)
```

```python
plt.plot(years, weights, c="y", lw=5, linestyle="--")
```

Bar plot

```python
x = ["C++", "C", "Python", "java", "C#"]
y = [30, 1, 70, 100, 50]
```

```python
plt.bar(x, y)
```

```python
plt.bar(years, weights)
```

```python
plt.bar(x, y, color="g", align="edge", width=0.5)
```

Histogram plot

```python
ages = np.random.normal(20, 1.5, 1000)
```

```python
plt.hist(ages)
```

```python
plt.hist(ages, bins=20)
```

```python
plt.hist(ages, bins=20, cumulative=True)
```

Pie chart

```python
langs = ["C++", "C", "Python", "java", "C#"]
votes = [30, 1, 70, 100, 50]
```

```python
plt.pie(votes, labels=langs)
```

```python
explodes = [0,0.2, 0,0,0]
```

```python
plt.pie(votes, labels=langs, explode=explodes, autopct = "%.2f%%", pctdistance=1.4, startangle=90)
```
